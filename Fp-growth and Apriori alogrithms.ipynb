```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
```

```python
def file_read(str):   
    new = open(str, "r")
    data = []
    for lines in new.readlines():
        temp = []
        for words in lines.split():
            temp.append(words)
        data.append(temp)
    return data

data1 = file_read("chess.dat")
data2 = file_read("mushroom.dat")
```

```python
transection = TransactionEncoder()
transection_array = transection.fit(data1).transform(data1)
df1 = pd.DataFrame(transection_array, columns=transection.columns_)
df1
```

```python
transection_array = transection.fit(data2).transform(data2)
df2 = pd.DataFrame(transection_array, columns=transection.columns_)
df2
```

```python
%timeit -n 5 -r 5 fpgrowth(df1, min_support=0.6)
%timeit -n 5 -r 5 apriori(df1, min_support=0.6)
```
